```python
# import neccessary packages
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# read raw data
raw_train = pd.read_csv("data/train.csv", parse_dates=['host_since', 'first_review', 'last_review'])
raw_test = pd.read_csv("data/test.csv", parse_dates=['host_since', 'first_review', 'last_review'])
```

```python
# Visualize the map of the properties
sns.scatterplot(data=raw_train, x='longitude', y='latitude', hue='price')
plt.title("Map of AirBnbs in the Training Dataset")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend(title="Price Category")
plt.show()
```

```python
# Examine the difference between the the different host listings count variables (and compare to y=x to see how they differ)
sns.scatterplot(data=raw_train, x='host_listings_count', y='host_total_listings_count', hue='price')
plt.plot([0, 1000], [0, 1000], color='black', linestyle='--', label='y = x')
plt.xlim(0, 500)  
plt.title("Host Listings vs. Host Total Listings (Internal Airbnb Metrics)")
plt.xlabel("Host Listings Count")
plt.ylabel("Total Listings Count")
plt.legend(title="Price Category")
plt.show()

sns.scatterplot(data=raw_train, x='host_total_listings_count', y='calculated_host_listings_count', hue='price')
plt.plot([0, 1000], [0, 1000], color='black', linestyle='--', label='y = x')
plt.xlim(0, 500)  
plt.title("Host Total Listings vs. Calculated Host Listings (NY Area)")
plt.xlabel("Host Total Listings Count")
plt.ylabel("Total Listings Count")
plt.legend(title="Price Category")
plt.show()
```
